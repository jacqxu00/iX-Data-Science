# Exercise. Using APIs:
We are going to use [teleport api](https://developers.teleport.org/api/), which is an api that among other things provide country and city information.

Using teleport's API, answer the question:

**Which country in Europe is the best one for an Architect in terms of salaries (50% percentile)?**

Make sure you check the API Explorer to find all of the required endpoints.



```python
import requests
from pprint import pprint
```

```python
# access the list of countries
url_countries = "https://api.teleport.org/api/continents/geonames%3AEU/countries/"
response = requests.get(url_countries)
countries = response.json()["_links"]["country:items"]
# pprint(countries)
```

```python
# go through each country and find 50% percentile salary of an architect
highest_salary = 0
best_country = ""

# go through each country
for country in countries:
    name = country["name"]
    url_salaries = country["href"] + "salaries"
    response2 = requests.get(url_salaries)
    salaries = response2.json()["salaries"]
    
    # find the 50% percentile salary of an architect
    for job in salaries:
        job_id = job["job"]["id"]
        job_salary_50 = job["salary_percentiles"]["percentile_50"]
        if job_id == "ARCHITECT":
            if job_salary_50 > highest_salary:
                highest_salary = job_salary_50
                best_country = name
            break

print(best_country + " (median salary: $%.2f)" %highest_salary)
```
